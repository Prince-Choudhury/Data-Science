```python
data1 = pd.DataFrame({'key1' : [1,2,4,5,6],
                      'key2' : [4,5,6,7,8],
                      'key3' : [3,4,5,6,6]
})
```

```python
data1
```

```python
data2 = pd.DataFrame({'key1' : [1,2,45,5,67],
                      'key4' : [56,5,6,7,8],
                      'key5' : [3,56,5,6,6]
})
```

```python
data2
```

```python
#Merge data1 and data2
pd.merge(data1, data2)   #By default inner merge : intersection of data1 and data2
```

```python
pd.merge(data1, data2, how = 'left')     #left merge : first all data come from data1 and then data2
```

```python
data1
```

```python
pd.merge(data1, data2, how = 'right')     #Right Merge : First data come from data2 and then data1
```

```python
pd.merge(data1, data2, how = 'outer')

#outer: use union of keys from both frames, similar to a SQL full outer
```

```python
pd.merge(data1, data2, how = 'cross')
```

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.
      
      #shift + tab

```python
pd.merge(data1, data2, how = 'outer', on = 'key1')      #to know more about on press shift+tab after data2,
```

```python
#convert Fare (price) euro into INR

df['Fare_INR'] = df['Fare'].apply(lambda x : x*80)
```

```python
df.head()
```

```python
#OR
def euro_inr(x):
    return x*80

df['Fare_INR'] = df['Fare'].apply(euro_inr)
```

```python
df.head()
```

```python
#Store Name length into another column
df['name_len'] = df['Name'].apply(len)
```

```python
df.head()
```

```python
#Create another column and put Fare inot cheap, mid, high category

def cat_fare(x):
    if x<10 : 
        return "cheap"
    elif x>=10 and x<20:
        return 'mid'
    else:
        return 'high'
```

```python
df['categ_fare'] = df['Fare'].apply(cat_fare)
```
