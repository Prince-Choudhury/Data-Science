```python
import pandas as pd

data = {'name': ['John', 'Jane', 'Bob'], 'age': [25, 30, 35]}

df = pd.DataFrame(data)

df['name_upper'] = df['name'].str.upper()
print(df)
```

```python
import pandas as pd

data = {'name': ['Alice', 'Bob', 'Charlie'], 'age': [30, 25, 40]}

df = pd.DataFrame(data)

max_age = df['age'].max()
print(max_age)
```

```python
import pandas as pd

data = {'date': ['2022-01-01', '2022-02-01', '2022-03-01'], 'sales': [100, 200, 300]}

df = pd.DataFrame(data)

df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.month
print(df)

#We're using the to_datetime() method to convert the 'date' column to a datetime format, 
#and then using the dt.month attribute to extract the month and assign it to a new column called 'month
```
