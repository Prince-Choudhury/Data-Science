```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
x
```

```python
y
```

```python
colors = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c = colors, s = sizes, alpha = .5 )   #alpha represents intensity of the colors
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```

```python
x = ['a', 'b', 'c', 'd', 'e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Plot")
```

```python
plt.barh(x,y)             #Horizontal Bar Plot
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Plot")
```

```python
x = np.linspace(0,10 , 200)
y = np.sin(x)
plt.figure(figsize = (6,2))   #We can control size of the graph
plt.plot(x,y, '--b')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my Title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c = '#BD33FF', s = sizes, alpha = .5 , marker='v')   #alpha represents intensity of the colors
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
#We can use HTML color codes
```

```python
x = [4,5,6,7,8,9]
y = [40,67,85,90,98,77]
colors = ['red', 'blue', '#BD33FF', '#5833FF', '#33FFEC', '#33FF64']
plt.scatter(x,y, c = colors)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z)
plt.show()
```
